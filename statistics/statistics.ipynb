```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
import pandas as pd

def p(x):

    return 1 / (np.abs(x)**2.2 + 1)

b = 1 / quad(p, -np.inf, np.inf)[0]
print(b)

def P(x, b):

    return b * p(x)

def E(x, b):

    return P(x, b) * x

quad(E, -np.inf, np.inf, args=(b,))

def E2(x, b):
    return P(x, b) * x**2

import war
print(quad(E2, -np.inf, np.inf, args=(b,), limit = 99999999, epsabs=1e-10, epsrel=1e-10, maxp1=500, limlst=500))
```
